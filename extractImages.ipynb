```python
# !pip install opencv-python

import cv2
import os
import glob

data_path = os.path.join(r"C:\Users\jing\Desktop\Data_enhancement\1.film\*mp4")
files = glob.glob(data_path)

print('cv2 version\t:',cv2.__version__) 

def extractImages(pathIn, pathOut):  

    vidcap = cv2.VideoCapture(pathIn)
    success,image = vidcap.read()
    print('image size\t:',image.shape[1],'X',image.shape[0],end='\n\n')  
    count = 0
    while success:
        vidcap.set(cv2.CAP_PROP_POS_MSEC,(count*200))   
        success,image = vidcap.read()                    
        print ('Read a new frame%d:'% count, success)          
        cv2.imwrite( pathOut + '\\frame%d.jpg' % count, image) 
        count = count + 1  
Video=files[0] #影片位置
Images=r'C:\Users\jing\Desktop\Data_enhancement\2.extractImages'  #圖片位置
extractImages(Video, Images)
```
